```python
# Divide the dataset in Independent Features and Dependent Feature
X = df.iloc[:,:-1] # Independent Features
y = df.iloc[:,-1] # Dependent Feature
```

```python
X
```

```python
y
```

```python
# Importing test train split 
from sklearn.model_selection import train_test_split
```

```python
# Spliting the independent and dependent feature into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=20)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Importing the clasifier and confusion matrix and accuracy score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Creating the estimator
log_reg = LogisticRegression()
```

```python
# Fit the data
log_reg.fit(X_train, y_train)
```

```python
# Prediction on test data
y_pred = log_reg.predict(X_test)
```

```python
y_pred
```

```python
# Getting the confusion matrix
confusion_matrix(y_pred, y_test)
```

```python
# Check the accuracy score
score = accuracy_score(y_pred, y_test)
```

```python
print(f'The accuracy score is: {round(score, 2)*100}%')
```
